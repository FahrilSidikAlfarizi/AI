```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import statsmodels.api as sm
from statsmodels.api import add_constant
import sklearn

```

```python
data=pd.read_csv('/content/placement.csv')
data
```

```python
x1=data['cgpa']
x1
```

```python
y=data['placement_exam_marks']
y
```

```python
plt.scatter(x1,y)
plt.xlabel('cgpa')
plt.ylabel('placement_exam_marks',fontsize=20)
plt.title('exam marks prediction',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
model = sm.OLS(y, x)
result = model.fit()
print(result.summary())
```

```python
plt.scatter(x1,y)
yhat=-0.8502*x1+38.1434
fig=plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('cgpa')
plt.ylabel('placement_exam_marks',fontsize=20)
plt.title('exam marks prediction',fontsize=20)
plt.show()
```
